```
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib as ml
import matplotlib.pyplot as plt

%matplotlib inline
ml.style.use('ggplot')
```

```
# Problem 1: Plot a rugplot on given below random generated data.
num = np.random.randn(170)
sns.rugplot(num,linewidth=3)
```

```
# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.
iris = sns.load_dataset('iris')
iris.head()
```

```
sns.pairplot(iris,hue='species')
plt.show()
```

**CONCLUSIONS :**

1. Going by the univariate plots of petal_length vs petal_length, then, petal_width vs petal_width, we can easily differentitate *iris-setosa* from the other two species.

2. The general trend in scatter plots is that, the data points for *iris-setosa* are always apart from the other two, which are always clustered together. So, we can conclude that, **the features of iris-setosa vary a lot from iris-virginica and iris-versicolor. Iris-virginica and iris-versicolor are almost similar in terms of the features, seen strongly for comparisons like sepal_length vs sepal_width.**

```
# Problem 3: Plot a joint plot showing relation between 'petal length' and 'petal width' with density estimates.

sns.jointplot(x='petal_length',y='petal_width',kind='kde',data=iris,color='orange')
plt.show()
```

```
# Problem 4: Plot a pie chart of survived based on sex.
titanic=sns.load_dataset('titanic')
titanic.head()
```

```
new1 = titanic[titanic.survived==1]
labels = ['Females','Males']
plt.pie(new1.sex.value_counts(),labels=labels,explode={0.1,0},autopct="%0.2f%%")
plt.title("PIE CHART SHOWING SURVIVAL RATIO BASED ON SEX")
plt.show()
```
